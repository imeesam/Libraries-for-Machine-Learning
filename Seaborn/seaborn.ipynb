#### seaborn line plot

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
## .load_dataset() is one of method of seaborn and we have a built in data set called "fmri"
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
## to make line plot b/w two column we have to use method .plotline() in which we have to pass two parameters x and y 
## we have an atribute in .plotline() called data which is basically im building a lineplot on a fmri data set
```

```python
sns.lineplot(x = "timepoint" , y = "signal" , data = fmri )
plt.show()
```

```python
### we have an new atribute called hue if we want to mapp a coulmn in plot we will use hue and give parameter(coulmn) to it 
```

```python
sns.lineplot(x = "timepoint" , y = "signal" , data = fmri , hue = "event" )
plt.show()
```

```python
### adding styles in it 

## style and hue are determined by the event column
```

```python
sns.lineplot(x = "timepoint" , y = "signal" , data = fmri , hue = "event" , style = "event" )
plt.show()
```

```python
## we can add markers in it just we have to set it to True 
```

```python
sns.lineplot(x = "timepoint" , y = "signal" , data = fmri , hue = "event" ,style = "event", markers = True )
plt.show()
```

#### seaborn histogram/distplot

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
### for histogramm using diamond dataframe 
```

```python
import pandas as pd
diamonds = pd.read_csv("diamonds.csv")
diamonds.head()
```

```python
## histogram aur distrubiton plot is the combination of frequency cruve and a histogram (histogram is used to understand the distribution of countinuos numerical value)
```

```python
## if we set his
```

```python
sns.distplot(diamonds["price"] , hist = False)
plt.show()
```

```python
## we ccan add color in it  iusing color attribute
```

```python
##  without hsit
sns.distplot(diamonds["price"] , hist = False , color = "red")
plt.show()
```

```python
##  with hist
sns.distplot(diamonds["price"] , color = "red")
plt.show()
```

```python
### now if i want distibution plot with frequency curve (i only want histo gram) then set kde = False

## and in it we can set bin = 10 
```

```python
sns.distplot(diamonds["price"] , kde = False , color = "red" , bins = 10)
plt.show()
```

```python
## to plot this above data  on diffrent axis use vetical = True
```

```python
sns.distplot(diamonds["price"] 
             , color = "red" , bins = 10 , vertical = True)
plt.show()
```

```python
## now we have a new geometry called .jointplot() its a combination of scatterplot and histogram
## creating this jointplot on iris data
```

```python
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
sns.jointplot(x = "SepalLengthCm" , y = "PetalLengthCm" , data =  iris)
plt.show()
```

```python
### you can give color in it with the help of attribuite
```

```python
sns.jointplot(x = "SepalLengthCm" , y = "PetalLengthCm" , data =  iris , color = "red")
plt.show()
```

```python
## if we want a regression line b/w scatter plot and histogram we have to use a new attribute called "kind"
## kind` must be one of ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']
```

```python
sns.jointplot(x = "SepalLengthCm" , y = "PetalLengthCm" , data =  iris , kind="reg")
plt.show()
```

```python
### next geometry is a box plot 
## we have to just use a box plot method and a new dataframe called churn this data frame tells about the diiferent features of different teleicom company
```

```python
churn = pd.read_csv('churn.csv')
churn.head()
```

```python
sns.boxplot(x = "Churn" , y = "tenure" , data = churn)
plt.show()
```

```python
sns.boxplot(x = "InternetService" , y = "MonthlyCharges" , data = churn)
plt.show()
```

```python
sns.boxplot(x = "Contract" , y = "tenure" , data = churn, palette = "Set1")
plt.show()
```

```python
## can increase the width by passing attribute of linewidth 
sns.boxplot(x = "Contract" , y = "tenure" , data = churn, linewidth = 3)
plt.show()
```

```python
### we can change the order of how these boxes are present ina very appropiate manner
### we have to give just order init.
```

```python
sns.boxplot(x = "Contract" , y = "tenure" , data = churn, order = ["Two year","One year","Month-to-month"] )
plt.show()
```

```python
sns.boxplot(x = "Contract" , y = "tenure" , data = churn, hue = "PaymentMethod" )
plt.show()
```
