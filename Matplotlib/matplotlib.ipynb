### matplotlib

```python
## matplotlib is libary which is used for data visulaization
```

```python
## starting from line plot

# .plot() --> takes two parameter and plot it 
# .xlabel()  --> Gives x label
# .ylabel()  --> Gives y label
# .title()   --> Gives the title
# .show()    --> shows the plot 
```

```python
import pandas as pd
import numpy as np ##  to genrate data 
from matplotlib import pyplot as plt   ### importing subplot pyplot 
```

```python
x = np.arange(1,11)
y = 2 * x
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
### for line style ::::   supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
plt.plot(x,y , color = 'g' , linewidth = 1 , linestyle = 'dotted') 
plt.title("Changing line asthetics")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
### adding two lines in the same plot 
```

```python
x = np.arange(1,11)
y1 = 2 * x
y2 = 4 *x
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1 , color = 'g' , linewidth = 2 , linestyle = 'dotted')
plt.plot(x,y2 , color = 'r' , linewidth = 2 , linestyle = 'dashed')
plt.title("ploting two lines in the same plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
## with using .grid()  method which basically shows the grid in plot 
```

```python
plt.plot(x,y1 , color = 'g' , linewidth = 2 , linestyle = 'dotted')
plt.plot(x,y2 , color = 'r' , linewidth = 2 , linestyle = 'dashed')
plt.title("ploting two lines in the same plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
### adding sub plots in the same plot using .subplot() method which basically creates two differnet plot on same section
### .subplot( rows ,columns  , index of plot)
```

```python
## when 1 rows 2 columns


plt.subplot(1,2,1)
plt.plot(x,y1 , color = 'g' , linewidth = 2 , linestyle = 'dotted')
plt.title("Subplot 1")
plt.xlabel("X-1")
plt.ylabel("Y-1")
plt.grid(True)


plt.subplot(1,2,2)
plt.plot(x,y2 , color = 'r' , linewidth = 2 , linestyle = 'dashed')
plt.title("Subplot 2")
plt.xlabel("X-2")
plt.ylabel("Y-2")
plt.grid(True)


plt.show()
```

```python
## when 2 rows 1 columns


plt.subplot(2,1,1)
plt.plot(x,y1 , color = 'green' , linewidth = 2 , linestyle = 'dotted')
plt.title("Subplot 1")
plt.xlabel("X-1")
plt.ylabel("Y-1")
plt.grid(True)


plt.subplot(2,1,2)
plt.plot(x,y2 , color = 'r' , linewidth = 2 , linestyle = 'dashed')
plt.title("Subplot 2")
plt.xlabel("X-2")
plt.ylabel("Y-2")
plt.grid(True)


plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1 ,color = 'green', linewidth = 5 , linestyle = "dashdot")
plt.grid(True)


plt.subplot(1,2,2)
plt.plot(x,y2 ,color = 'blue', linewidth = 2 , linestyle = "--")
plt.grid(True)

plt.show()
```

### barplot 

```python
### bar plot is used for catagorical values on x it must be names and on y there is numerical entity
## ----> .bar() is used for ploting catagoricla values this is for vetical bar
```

```python
## creating bar plot with the help of dictionary  by extracting values from it and ploting key and values
student = {"Bob": 87 , "Ali" : 49 , "Meesu" :90 }
```

```python
names = list(student.keys())
```

```python
marks =  list(student.values())
```

```python
plt.bar(names, marks)
plt.show()
```

#### adding title and labels in the plot 

```python
plt.bar(names,marks)
plt.title("First vertical Bar plot with 'title and labels'")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
## for horizontal bar use .barh()
```

```python
plt.barh(names, marks ,color = 'green')
plt.title("First horizontal Bar plot with 'title and labels'")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

### Scatter Plot

```python
## scatter plot are which is used to show data points in plot 
## is used to understand the diiference b/w two entties
## .scatter() is used to show the scatter plot takes two parameter x and y axes
```

```python
x = np.arange(10,100,10)
a = [13 ,7, 86, 45 ,8,29,50,34,79 ] 
```

```python
plt.scatter(x,a)
plt.show()
```

```python
### changing aesthetics of the plot 
```

```python
plt.scatter(x, a , marker ="*" , c = "red" , s = 100 )
plt.show()
```

```python
## adding two markers in same plot
x = np.arange(10, 100 , 10)
z = [ 50, 50, 50,50,50,50,50,50,50]
y1 = [ 10, 20, 30, 40, 50, 60, 70 , 80, 90 ]
y2 = [ 90, 80, 70, 60, 50, 40,30 , 20, 10 ]
plt.scatter(x , y1 , marker ="*" , c = "green" , s = 200 )
plt.scatter(x , y2 , marker ="*" , c = "green" , s = 200 )
plt.scatter(z , y2 , marker ="*" , c = "Red" , s = 200 )
plt.scatter(y1 , z , marker ="*" , c = "Red" , s = 200 )

plt.show()
```

```python
## subplots in scatter be like
```

```python
x = np.arange(10, 100 , 10)
z = [ 50, 50, 50,50,50,50,50,50,50]
y1 = [ 10, 20, 30, 40, 50, 60, 70 , 80, 90 ]
y2 = [ 90, 80, 70, 60, 50, 40,30 , 20, 10 ]

plt.subplot(1 , 2 ,1)
plt.scatter(x , y1 , marker ="*" , c = "green" , s = 200 )
plt.title("Subplot 1")
plt.xlabel("X-1")
plt.ylabel("Y-1")
plt.grid(True)

plt.subplot(1 , 2 ,2)
plt.scatter(x , y2 , marker ="*" , c = "green" , s = 200 )
plt.title("Subplot 2")
plt.xlabel("X-2")
plt.ylabel("Y-2")
plt.grid(True)

plt.show()
```

#### pie plot


```python
## we can use either dictonary and extract keys and values or either create 2 list and go on its your choice
```

```python
dt = { "Apple" : 10 , "Mango" : 40 , "Grapes" : 30 , "banana" : 25 }
```

```python
plt.pie(dt.values() , labels = dt.keys())
plt.plot()
```

```python
### we can change asthetics of piechart by following things
```

```python
### to add the actual % we will use autopct = '%0.1%%' 
## to use autopct you have to give this value in '' single quatoes 
```

```python
## to add color and % we gonna use above code again
## and to use color in pie use atrbiute colors = [] it will follow indexing 
```

```python
dt = { "Apple" : 10 , "Mango" : 40 , "Grapes" : 30 , "banana" : 25 }
plt.pie(dt.values() , labels = dt.keys() , autopct = '%0.1f%%' , colors = ["yellow","gray","blue","black" ])
plt.plot()
```

### doughNut - chart




```python
### to use doughNut you have to use .pie() two times one as usual outer one it will have radius greater than inner and other as inner which will be white
```

```python
plt.pie(dt.values() , labels = dt.keys() , radius = 1)
plt.pie([1] , colors = ['w'], radius = 0.6)
plt.show()
```
